```python
def Environment(n):
    ##Random initialization of rubbish in a n*n sea grid
    np.random.seed(7) #for reproducible results
    rubbish_initialization = np.random.randint(0, 5, size=(n, n))
    sea_creatures_initialization = np.random.randint(0, 3, size=(n,n))
    #print pd.DataFrame(np.reshape(np.asarray(rubbish_initialization), (25,1)),columns=['rubbish'])
    rubbish_sea_grid = pd.DataFrame(rubbish_initialization)
    sea_creature_sea_grid = pd.DataFrame(sea_creatures_initialization)
    print "rubbish:\n",rubbish_sea_grid
    print "sea_creature:\n",sea_creature_sea_grid
    #print np.reshape(np.asarray(rubbish_initialization), (25,1))
    ##Storing dataset in pandas framework
    #Getting list of x and y for location point storage
    x = [0] *n
    for i in range (1,n):
        x1 = [i]*n
        x = np.concatenate((x,x1), axis=0)
    #print x

    y1 = np.arange(n)
    y = y1
    for i in range(1,n):
        y = np.concatenate((y,y1), axis=0)
    #print y

    location = {'location_x': x, 'location_y': y}
    df1 = pd.DataFrame(data=location)
    #print df1
    df2 = pd.DataFrame(np.reshape(np.asarray(rubbish_initialization), (n*n,1)),columns=['rubbish'])
    #print df2
    df3 = pd.DataFrame(np.reshape(np.asarray(sea_creatures_initialization), (n*n,1)),columns=['sea_creatures'])
    df = df1.join(df2)
    df = df.join(df3)
    print df
    return df
```

```python
def wind_blowing_mech(wind_prob, n, df):
    ##Wind blowing Mechanism
    #50 percent chance of wind belowing
    if np.random.rand() < wind_prob:
        print "Blow Wind Blow!"
        #Equal chance of Wind Blowing in four directions: left,up,right,down 
        wind_dir_prob = np.random.rand()
        if  wind_dir_prob < 0.25:
            wind_dir = 'left'
        elif wind_dir_prob < 0.5:
            wind_dir = 'up'
        elif wind_dir_prob <0.75:
            wind_dir = 'right'
        else:
            wind_dir = 'down'

        #Wind Blows Rubbish a square to the the direction of wind, rubbish gets collated if it is at the bounds
        if wind_dir == 'left':
            print 'left!'
            temp = df.copy()
            for i in range(0,n*n,n):
                for j in range(n):
                    if j == 0:
                        df.rubbish[i+j] = df.rubbish[i+j] + temp.rubbish[i+j+1]
                    elif j == (n-1):
                        df.rubbish[i+j] = 0
                    else:
                        df.rubbish[i+j] = temp.rubbish[i+j+1]
            print df

        if wind_dir == 'right':
            print 'right!'
            temp = df.copy()
            for i in range(0, n*n, n):
                for j in range(n):
                    if j == 0:
                        df.rubbish[i+j] = 0
                    elif j == (n-1):
                        df.rubbish[i+j] = df.rubbish[i+j] + temp.rubbish[i+j-1]
                    else:
                        df.rubbish[i+j] = temp.rubbish[i+j-1]
            print df


        if wind_dir == 'up':
            print 'up!'
            temp = df.copy()
            for i in range(n):
                for j in range(i,n*n,n):
                    if j > ((n-1)*n-1):
                        df.rubbish[j] = 0  #bottom column
                    elif j < n:
                        df.rubbish[j] = df.rubbish[j] + temp.rubbish[j+n]
                    else:
                        df.rubbish[j] = temp.rubbish[j+n]
            print df


        if wind_dir == 'down':
            print 'down!'
            temp = df.copy()
            for i in range(n):
                for j in range(i,n*n,n):
                    if j < n:
                        df.rubbish[j] = 0            #most left column
                    elif j > ((n-1)*n-1):
                        df.rubbish[j] = df.rubbish[j] + temp.rubbish[j-n]
                    else:
                        df.rubbish[j] = temp.rubbish[j-n]
            print df
    else:
        print "no wind!"
    return df
```
